```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import MinMaxScaler
```

# New section

```python
df=pd.read_csv('/content/car_purchasing.csv',encoding='ISO-8859-1')
df
```

```python
df.columns
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(columns=['customer name','customer e-mail', 'country'],inplace=True)
```

```python
df.columns
```

```python
x=df.drop(columns=['car purchase amount'])
y=df[['car purchase amount']]
```

Scalling

```python
le=MinMaxScaler()
X=le.fit_transform(x)
Y=le.fit_transform(y)
X
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,shuffle=True,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
y_pre=model.predict(X_test)
```

```python
print(mean_squared_error(Y_test,y_pre))
```
